```python
import tkinter as tk
from tkinter import Label
from PIL import Image, ImageTk
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import joblib
from cvzone.HandTrackingModule import HandDetector

# Load the trained model and label binarizer
model = load_model("hand_gesture_model_A_Z.h5")
lb = joblib.load("label_binarizer.pkl")

# Initialize HandDetector
detector = HandDetector(maxHands=1)

# Tkinter setup
root = tk.Tk()
root.title("Hand Gesture Recognition")

# Create a label to display the video feed
label = Label(root)
label.pack()

def update_frame():
    # Capture frame-by-frame
    success, img = cap.read()
    if not success:
        return

    # Detect hands
    hands, img = detector.findHands(img)
    if hands:
        hand = hands[0]
        x, y, w, h = hand['bbox']
        
        # Ensure bounding box stays within image bounds
        x = max(0, x - 20)
        y = max(0, y - 20)
        w = min(img.shape[1] - x, w + 40)
        h = min(img.shape[0] - y, h + 40)
        
        imgCrop = img[y:y + h, x:x + w]  # Adjusted cropping
        
        # Check if the cropped image has valid dimensions
        if imgCrop.shape[0] == 0 or imgCrop.shape[1] == 0:
            return
        
        imgResize = cv2.resize(imgCrop, (300, 300))
        imgResize = np.expand_dims(imgResize, axis=0) / 255.0
        
        # Predict
        pred = model.predict(imgResize)
        pred_class = lb.inverse_transform(pred)[0]
        cv2.putText(img, pred_class, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # Convert the image to RGB format for Tkinter
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_pil = Image.fromarray(img_rgb)
    img_tk = ImageTk.PhotoImage(image=img_pil)
    
    # Update Tkinter label
    label.config(image=img_tk)
    label.image = img_tk  # Keep a reference to avoid garbage collection
    
    # Update the frame
    root.after(10, update_frame)

# Open video capture
cap = cv2.VideoCapture(0)

# Start updating frames
update_frame()

# Start the Tkinter main loop
root.mainloop()

# Release the video capture object
cap.release()
```
